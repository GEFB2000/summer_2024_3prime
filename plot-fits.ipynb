### Meta samples

**1st: Set dirs**

```python
# when working with meta samples
# celltype = 'HEp2'
celltype = 'foreskin_fibroblast'
gene_of_interest = 'RPS6KC1|NM_001349668.2'
```

```python
workdir = "/scratch/Users/geba9152/LIET-summer2024/meta-celltype-comparison/"
config_file = workdir+f"config/tight-sT/meta_{celltype}.liet.config"
results_file = workdir+f"results/tight-sT/meta_{celltype}.liet"
log_file = workdir+f"results/tight-sT/meta_{celltype}.liet.log"

# results_loader() ---> {'gene_ID': (xvals, preads, nreads, strand, model_params), ...}
gene_list = [gene_of_interest]
gene_results = results_loader(gene_list, config=config_file, result=results_file, log=log_file)
```

**2nd: plot**

```python
fig, ax = plt.subplots(1, 1, figsize=[10, 6])

xvals, preads, nreads, strand, model_params = gene_results[gene_of_interest]
    
rp.LIET_ax(ax, **model_params, strand = strand, xvals = xvals, data = (preads, nreads), nbins = 100)

plt.ylim([-1.0e-4, 1.0e-4])
plt.ylabel('Probability');
plt.xlabel(f'Genomic Coordinate relative to {gene_of_interest} TSS (bp)');

plt.show()
```
